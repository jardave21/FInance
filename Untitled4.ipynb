```python
def calculate_z_score(lq, sq, roa, ebit, tev):
    z_score = 1.2*lq + 1.4*sq + 3.3*roa + 0.6*ebit + 1.0*tev
    return z_score

# Ejemplo de c√°lculo del z-score de una empresa
lq = 0.06  # coeficiente de liquidez
sq = 0.01  # coeficiente de solvencia
roa = 0.01  # coeficiente de rentabilidad sobre activos
ebit = 1.07  # coeficiente de rentabilidad operativa
tev = 1.08  # coeficiente de apalancamiento financiero

z_score = calculate_z_score(lq, sq, roa, ebit, tev)
print("El z-score de la empresa es:", z_score)
```
