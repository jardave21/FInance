```python
#Importamos las librerias
from backtesting import Backtest, Strategy
from backtesting.lib import crossover,plot_heatmaps

import yfinance as yf

import numpy
import talib
import pandas as pd
```

```python
#Descargamos los datos
#data = yf.download("BTC-USD",start="2012-01-01",end="2022-01-01")
data = yf.download("XRP-BTC",start="2016-01-01",end="2022-01-01")
#data = yf.download("XRP-USD",start="2016-01-01",end="2022-01-01")
```

```python
#Visualizamos los datos
data
```

```python
#Creamos la estrategia y el backtest
class SMAcross(Strategy):
  ema_chica= 9 
  ema_grande= 55

  def init(self):

    self.smaQuick = self.I(talib.SMA,self.data.Close,self.ema_chica)
    self.smaSlow = self.I(talib.SMA,self.data.Close,self.ema_grande)

  def next(self):
    price = self.data.Close[-1]
    if crossover(self.smaQuick,self.smaSlow):
      self.buy(sl = price*0.95,tp = price*1.1)
    elif crossover(self.smaSlow,self.smaQuick):
      self.sell(sl = price*1.1,tp = price*0.95)

bt = Backtest(data,SMAcross,exclusive_orders=True)
```

```python
#Estrategia de comprar y vender
class SimpleBuy(Strategy):
  
  def init(self):
    self.od = None

  def next(self):
    price = self.data.Close[-1]
    
    if self.orders:
      return
    
    self.od = self.buy(limit = price * 1.05, sl = price,tp = price*1.1)

bt = Backtest(data,SimpleBuy,cash = 100000)
```

```python
#Estrategia RSI
class RSIcross(Strategy):
  
  rsi_period = 14
  slPercent = 0.95

  def init(self):
    self.rsi = self.I(talib.RSI,self.data.Close,self.rsi_period)

  def next(self):
    price = self.data.Close[-1]

    if self.rsi[-1] < 30:
      self.position.close()

    if self.position.is_long:
      return
    
    
    self.buy(sl = price * self.slPercent)

bt = Backtest(data,RSIcross,cash = 100000)
```

```python
#Ejecutamos el backtest
stats = bt.run()

#Vamos los resultados del backtest
stats
```

```python
#Grafico con los resultados
bt.plot()
```
