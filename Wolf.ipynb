```python
# Importar las librerías necesarias
import time
from binance.client import Client

# Crear un cliente de Binance
client = Client()

while True:
    # Realiza una llamada a la API para obtener información sobre el par BTC/USDT
    ticker = client.get_ticker(symbol='BTCUSDT')

    # Imprime la información obtenida
    #print(ticker)

    # Abrir un archivo de texto en modo escritura
    with open("precios_binance.txt", "a") as f:
      # Escribir el símbolo y el precio en una línea del archivo
      f.write(str(ticker) + "\n")
    # Espera un segundo antes de hacer la siguiente llamada a la API
    time.sleep(1)
```
