```python
import yfinance as yf
import matplotlib.pyplot as plt

# Definir los símbolos de las criptomonedas y el rango de fechas
symbols = ["BTC-USD"]
start_date = "2018-01-01"
end_date = "2023-03-22"

# Descargar los precios usando yfinance
crypto_prices = yf.download(symbols, start=start_date, end=end_date)

# Seleccionar sólo la columna de "Close" (precio de cierre)
crypto_prices = crypto_prices["Close"]

# Graficar los precios de las criptomonedas
crypto_prices.plot(figsize=(10, 7))
plt.xlabel("Fecha")
plt.ylabel("Precio en USD")
plt.title("Precios de Bitcoin, Ethereum y XRP desde 2018")
plt.legend(symbols)
plt.show()
```

```python

import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

# Definir los símbolos de las criptomonedas y el rango de fechas
symbols = ["BTC-USD", "ETH-USD", "XRP-USD"]
start_date = "2018-01-01"
end_date = "2022-01-01"

# Descargar los precios usando yfinance
crypto_prices = yf.download(symbols, start=start_date, end=end_date)

# Seleccionar sólo la columna de "Close" (precio de cierre)
crypto_prices = crypto_prices["Close"]

# Calcular la matriz de correlación entre los precios de cierre de cada activo
corr_matrix = crypto_prices.corr()

# Graficar la matriz de correlación
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlación entre precios de criptomonedas")
plt.show()
```
