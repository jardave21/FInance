```python
import numpy as np
import matplotlib.pyplot as plt

def update_board(board):
    # Calculate number of neighbours for each cell
    nbrs_count = np.zeros(board.shape)
    for r in [-1, 0, 1]:
        for c in [-1, 0, 1]:
            if not (r == 0 and c == 0):
                nbrs_count += np.roll(np.roll(board, r, 0), c, 1)

    # Apply rules of the game
    born = (board == 0) & (nbrs_count == 3)
    survive = ((board == 1) & (nbrs_count == 2)) | ((board == 1) & (nbrs_count == 3))
    board[born | survive] = 1
    board[~(born | survive)] = 0
    return board

def run_game(board, num_iters):
    for i in range(num_iters):
        board = update_board(board)
    return board

# Set up an initial board
board = np.zeros((50, 50))
board[10:15, 10:15] = 1
board[35:40, 35:40] = 1

# Run the game
while True: 
  final_board = run_game(board, num_iters=10)

# Plot the final board
  plt.imshow(final_board, cmap='Greys')
  plt.show()
```
